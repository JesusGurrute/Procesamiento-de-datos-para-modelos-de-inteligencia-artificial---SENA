**9.** A un trabajador le pagan seg√∫n sus horas y una tarifa de pago por horas. Si la cantidad de horas trabajadas es mayor a 40 horas. La tarifa se incrementa en un 50% para las horas extras. Calcular el salario del trabajador dadas las horas trabajadas y la tarifa.

```python
tarifa_hora = float(input("Ingrese la tarifa por hora de trabajo: "))
horas = float(input("Ingrese el n√∫mero de horas trabajadas: "))
porcentaje_extra = float(input("Ingrese el porcentaje extra para horas extra (ej. 50): "))

if horas <= 0:
    print("El trabajador no recibe pago")
else:
    if horas <= 40:
        salario_total = tarifa_hora * horas
        print(f"El trabajador recibe ${salario_total:.2f}")
    else:
        horas_extra = horas - 40
        tarifa_extra = tarifa_hora * (1 + porcentaje_extra / 100)

        salario_normal = tarifa_hora * 40
        salario_extra = tarifa_extra * horas_extra
        salario_total = salario_normal + salario_extra

        print(f"Se trabajaron {horas_extra} horas extra con tarifa de ${tarifa_extra:.2f}")
        print(f"El trabajador recibe ${salario_normal:.2f} + ${salario_extra:.2f} = ${salario_total:.2f}")
```

**10.** Dada la duraci√≥n en minutos de una llamada calcular el costo, considerando: Hasta tres minutos el costo es \$ 50 por minuto. Por encima de tres minutos es \$50 menos \$ 10 por cada minuto adicional.


```python
duracion = float(input("Duraci√≥n de llamada: "))
costo_base = 50

if duracion <= 0:
    print("La llamada no tiene costo")
elif duracion <= 3:
    costo_total = duracion * costo_base
    print(f"La llamada tiene un costo de ${costo_total:.2f}")
else:
    minutos_adicionales = duracion - 3
    costo_total = (3 * costo_base) + (minutos_adicionales * (costo_base - 10))
    print(f"La llamada tiene un costo de ${costo_total:.2f}")
```

**12.** Escribir un algoritmo que determine si un a√±o es bisiesto. Un a√±o es bisiesto si es m√∫ltiplo de 4 (por ejemplo 1984). Los a√±os m√∫ltiplos de 100 no son bisiestos, salvo si ellos son tambi√©n m√∫ltiplos de 400 (2000 es bisiesto, pero; 1800 no lo es).

```python
anio = int(input("Ingrese un a√±o: "))

if 1 <= anio <= 9999:
    if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
        print(f"{anio} es bisiesto")
    else:
        print(f"{anio} no es bisiesto")
else:
    print("Los a√±os permitidos van desde el a√±o 1 hasta el a√±o 9999")
```
